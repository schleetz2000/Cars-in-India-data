```python
import pandas as pd
```

```python
# Loading data from the CSV file to panda framework

df = pd.read_csv('/content/car_details.csv')
```

```python
# Dropping lines with null value

df.dropna(inplace=True)
```

```python
# Treating data about the engines

df['Engine'] = df['Engine'].str.replace(' cc', '').astype(float)
```

```python
df["Location"].unique()
```

```python
# Calculating the average price per color

avrg_average_price_color = (df.groupby('Color')['Price'].mean()).round(2)

# Converting to brazillian reais

avrg_average_price_color = (avrg_average_price_color.divide(17)).round(2)
```

```python
# Calculating the average price per brand

avrg_price_for_brand = df.groupby('Make')['Price'].mean()

# Converting to brazillian reais

avrg_price_for_brand = (avrg_price_for_brand.divide(17)).round(2)
```

```python
# Calculating the quantity of cars per brand

qtty_cars_per_brand = df.groupby('Make').size()
```

```python
# Calculating the quantity of cars by type of transmission

qtty_cars_per_transmission = df.groupby('Transmission').size()
```

```python
# Calculate the average age per cars in the

avrg_age_per_region = (df.groupby('Location')['Year'].mean()).astype(int)
```

```python
# Calculate the average engine power per type of fuel use

avrg_power_per_fuel = (df.groupby('Fuel Type')['Engine'].mean()).round(2)
```

```python
print("\n Average car price per color:")
print(avrg_average_price_color.sort_values(ascending=False))
```

```python
print("\n Average car price per brand:")
print(avrg_price_for_brand.sort_values(ascending=False))
```

```python
print("\n Quantity of cars per brand:")
print(qtty_cars_per_brand.sort_values(ascending=False))
```

```python
print("\n Quantity of cars by type of transmission:")
print(qtty_cars_per_transmission.sort_values(ascending=False))
```

```python
print("\n Average age of cars per region:")
print(avrg_age_per_region.sort_values(ascending=False))
```

```python
print("\n average engine power per type of fuel used:")
print(avrg_power_per_fuel.sort_values(ascending=False))
```
